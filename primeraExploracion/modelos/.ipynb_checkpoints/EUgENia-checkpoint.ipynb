```python
import os
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import cv2

import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.graph_objects as go

import matplotlib.pyplot as plt
```

```python
inventario = pd.read_csv('/home/ricmunrom/Documents/jorgeLozano/EUgeNia/primeraExploracion/inputsSinteticos/inventario.csv')
demandaEsperada = pd.read_csv('/home/ricmunrom/Documents/jorgeLozano/EUgeNia/primeraExploracion/inputsSinteticos/demandaEsperada.csv')
ordenesCompra = pd.read_csv('/home/ricmunrom/Documents/jorgeLozano/EUgeNia/primeraExploracion/inputsSinteticos/ordenesCompra.csv')
CEDI = pd.read_csv('/home/ricmunrom/Documents/jorgeLozano/EUgeNia/primeraExploracion/inputsSinteticos/CEDI.csv')
```

###### Que hace el codigo?

Implementa la reducción diaria del inventario en tienda según la demanda esperada de los artículos. Además, se registran los movimientos de reabastecimiento de mercancía desde el CEDI hacia las tiendas y los movimientos de recepción de mercancía por el CEDI. Se guarda el historial de todas las operaciones en los DataFrames historial_inventario e historial_CEDI.

Creamos un nuevo dataFrame llamado CEDI_alDia_df que va actualizando el nivel de inventario del ultimo dia observado en el dataFrame CEDI segun los movimientos registrados en historial_CEDI_df.

Ademas permite hacer transferencia de mercancias entre tiendas

###### Nota Importante


1. La Resta es con Inventario Antes y no Con Inventario Despues


2. Hay unas fechas que se tienen que cambiar manual si se actualiza la informacion sintetica

```python
# Convertir las columnas de fecha a tipo datetime
demandaEsperada['Fecha'] = pd.to_datetime(demandaEsperada['Fecha'])
ordenesCompra['Fecha Orden de Compra'] = pd.to_datetime(ordenesCompra['Fecha Orden de Compra'])
ordenesCompra['Fecha Recepcion Mercancia'] = pd.to_datetime(ordenesCompra['Fecha Recepcion Mercancia'])

# Crear DataFrame para registrar los cambios diarios en el inventario
historial_inventario = []

# Inicializar inventario diario de cada tienda
inventario_diario = inventario[['SKU', 'Tienda', 'Unidades en Inventario']].copy()
inventario_diario.columns = ['SKU', 'Tienda', 'Inventario']

# Crear un diccionario para rastrear el inventario remanente
inventario_remanente = inventario_diario.set_index(['SKU', 'Tienda'])['Inventario'].to_dict()

# Crear un diccionario para el inventario del CEDI
inventario_CEDI = CEDI.set_index('SKU')['Unidades Disponible'].to_dict()

# Crear un DataFrame para registrar los envíos del CEDI
historial_CEDI = []

# Crear un DataFrame para registrar los envíos entre tiendas
historial_envios_tiendas = []

# Filtrar las órdenes de compra con fecha de orden mayor o igual a 2024-05-13
ordenes_filtradas = ordenesCompra[ordenesCompra['Fecha Orden de Compra'] >= '2024-05-13']

# Iterar sobre las órdenes filtradas para actualizar el inventario del CEDI
for _, row in ordenes_filtradas.iterrows():
    sku = row['SKU']
    unidades_recibidas = row['Volumen Recibido']
    fecha_recepcion = row['Fecha Recepcion Mercancia']
    
    # Sumar las unidades recibidas al inventario del CEDI
    if sku in inventario_CEDI:
        inventario_CEDI[sku] += unidades_recibidas
    else:
        inventario_CEDI[sku] = unidades_recibidas
    
    # Registrar la recepción en el historial del CEDI
    historial_CEDI.append({
        'Fecha': fecha_recepcion,
        'SKU': sku,
        'Tienda': 'CEDI',
        'Unidades Recibidas': unidades_recibidas,
        'Unidades Enviadas': 0  # Asegurarse de que esta columna exista
    })

# Estas fechas requieren de un ajuste manual

# Continuar con la simulación diaria del inventario
fechas = pd.date_range(start='2024-05-20', end='2024-06-20')

for fecha in fechas:
    # Filtrar la demanda esperada para el día actual
    demanda_dia = demandaEsperada[demandaEsperada['Fecha'] == fecha]
    
    # Reabastecer desde el CEDI antes de abrir la tienda
    for _, row in demanda_dia.iterrows():
        sku = row['SKU']
        tienda = row['Tienda']
        demanda = row['Demanda Esperada']
        
        # Verificar si el inventario remanente es insuficiente
        if inventario_remanente[(sku, tienda)] < demanda:
            # Calcular la cantidad a reabastecer basada en la demanda esperada
            cantidad_reabastecer = min(inventario_CEDI.get(sku, 0), demanda - inventario_remanente[(sku, tienda)])
            if cantidad_reabastecer > 0:
                # Actualizar el inventario remanente
                inventario_remanente[(sku, tienda)] += cantidad_reabastecer
                
                # Actualizar el inventario del CEDI
                inventario_CEDI[sku] -= cantidad_reabastecer
                
                # Registrar el envío en el historial del CEDI
                historial_CEDI.append({
                    'Fecha': fecha,
                    'SKU': sku,
                    'Tienda': tienda,
                    'Unidades Recibidas': 0,  # Asegurarse de que esta columna exista
                    'Unidades Enviadas': cantidad_reabastecer
                })
                
            # Si aún hay demanda insatisfecha, intentar transferir desde otra tienda
            if inventario_remanente[(sku, tienda)] < demanda:
                tiendas_con_exceso = [(t, inv) for (s, t), inv in inventario_remanente.items() if s == sku and t != tienda and inv > demanda]
                for tienda_exceso, inventario_exceso in tiendas_con_exceso:
                    if inventario_remanente[(sku, tienda)] >= demanda:
                        break
                    cantidad_transferir = min(inventario_exceso, demanda - inventario_remanente[(sku, tienda)])
                    if cantidad_transferir > 0:
                        # Actualizar inventarios de las tiendas
                        inventario_remanente[(sku, tienda)] += cantidad_transferir
                        inventario_remanente[(sku, tienda_exceso)] -= cantidad_transferir
                        
                        # Registrar el envío entre tiendas
                        historial_envios_tiendas.append({
                            'Fecha': fecha,
                            'SKU': sku,
                            'Tienda Origen': tienda_exceso,
                            'Tienda Destino': tienda,
                            'Unidades Transferidas': cantidad_transferir
                        })
    
    # Actualizar inventario remanente con demanda esperada del día actual
    for _, row in demanda_dia.iterrows():
        sku = row['SKU']
        tienda = row['Tienda']
        demanda = row['Demanda Esperada']
        
        # Calcular inventario antes y después de la demanda
        inventario_actual = inventario_remanente[(sku, tienda)]
        nuevo_inventario = max(0, inventario_actual - demanda)
        
        # Registrar el inventario antes y después de la reducción
        historial_inventario.append({
            'Fecha': fecha,
            'SKU': sku,
            'Tienda': tienda,
            'Inventario Antes': inventario_actual,
            'Inventario Despues': nuevo_inventario
        })
        
        # Actualizar el inventario remanente
        inventario_remanente[(sku, tienda)] = nuevo_inventario

# Convertir los historiales a DataFrames
historial_inventario_df = pd.DataFrame(historial_inventario)
historial_CEDI_df = pd.DataFrame(historial_CEDI)
historial_envios_tiendas_df = pd.DataFrame(historial_envios_tiendas)

# Guardar los outputs
historial_inventario_df.to_csv('/home/ricmunrom/Documents/jorgeLozano/EUgeNia/primeraExploracion/outputsModelo/historialInventario.csv', index=False)
historial_CEDI_df.to_csv('/home/ricmunrom/Documents/jorgeLozano/EUgeNia/primeraExploracion/outputsModelo/historialCEDI.csv', index=False)
historial_envios_tiendas_df.to_csv('/home/ricmunrom/Documents/jorgeLozano/EUgeNia/primeraExploracion/outputsModelo/historialEnviosTiendas.csv', index=False)

# Convertir las fechas a formato datetime para facilitar las operaciones
CEDI['Fecha'] = pd.to_datetime(CEDI['Fecha'])
historial_CEDI_df['Fecha'] = pd.to_datetime(historial_CEDI_df['Fecha'])

# Combinar las unidades iniciales con el historial
CEDI_inicial_df = CEDI[['Fecha', 'SKU', 'Unidades Disponible']].copy()
CEDI_inicial_df.rename(columns={'Unidades Disponible': 'Unidades Cambios'}, inplace=True)

historial_CEDI_df['Unidades Cambios'] = historial_CEDI_df['Unidades Recibidas'].fillna(0) - historial_CEDI_df['Unidades Enviadas'].fillna(0)
historial_CEDI_df.sort_values(by='Fecha', inplace=True)
historial_CEDI_df = historial_CEDI_df[['Fecha', 'SKU', 'Tienda', 'Unidades Cambios']]  # Mantener columna 'Tienda'

# Concatenar los DataFrames
CEDI_combined_df = pd.concat([CEDI_inicial_df, historial_CEDI_df], ignore_index=True)

# Ordenar por SKU y Fecha
CEDI_combined_df.sort_values(['SKU', 'Fecha'], inplace=True)

# Calcular las unidades disponibles acumuladas
CEDI_combined_df['Unidades Disponible'] = CEDI_combined_df.groupby('SKU')['Unidades Cambios'].cumsum()

# Filtrar solo las filas necesarias
CEDI_alDia_df = CEDI_combined_df[['Fecha', 'SKU', 'Tienda', 'Unidades Disponible']].copy()

# Calcular la diferencia entre inventario y demanda esperada para el último día del periodo
fecha_final = '2024-06-20'
inventario_final_dia = historial_inventario_df[historial_inventario_df['Fecha'] == fecha_final].copy()
demanda_final_dia = demandaEsperada[demandaEsperada['Fecha'] == fecha_final].copy()

# Combinar los DataFrames para calcular la diferencia
diferencia_inventario_df = inventario_final_dia.merge(demanda_final_dia, on=['SKU', 'Tienda'], suffixes=('_Inventario', '_Demanda'))

# Calcular la diferencia
diferencia_inventario_df['Diferencia'] = diferencia_inventario_df['Inventario Antes'] - diferencia_inventario_df['Demanda Esperada']

# Guardar el DataFrame de diferencias
diferencia_inventario_df.to_csv('/home/ricmunrom/Documents/jorgeLozano/EUgeNia/primeraExploracion/outputsModelo/diferenciaInventario.csv', index=False)
```

###### Visualizaciones

Para la visualizacion a nivel diario de demanda e inventario por tienda se utiliza historial_inventario_df y demandaEsperada

```python
# Suponiendo que historial_inventario_df y demandaEsperada ya están definidos y cargados
# Aquí se mostrarán las columnas necesarias

# Initialize the Dash app
app = dash.Dash(__name__)

# Get unique stores and dates
unique_stores = historial_inventario_df['Tienda'].unique()
unique_dates = pd.to_datetime(historial_inventario_df['Fecha'].unique())

# Layout of the Dash app
app.layout = html.Div([
    dcc.Dropdown(
        id='store-dropdown',
        options=[{'label': store, 'value': store} for store in unique_stores],
        value=unique_stores[0],
        placeholder="Select a store"
    ),
    dcc.Graph(id='inventory-demand-graph'),
    dcc.Slider(
        id='date-slider',
        min=unique_dates.min().timestamp(),
        max=unique_dates.max().timestamp(),
        value=unique_dates.min().timestamp(),
        marks={date.timestamp(): date.strftime('%Y-%m-%d') for date in unique_dates},
        step=86400
    ),
    html.Div(id='difference-marker', style={'font-size': '24px', 'margin-top': '20px'})
])

# Callback to update the graph and difference marker based on the selected store and date
@app.callback(
    [Output('inventory-demand-graph', 'figure'),
     Output('difference-marker', 'children')],
    [Input('store-dropdown', 'value'),
     Input('date-slider', 'value')]
)
def update_graph_and_marker(selected_store, selected_date):
    selected_date = pd.to_datetime(selected_date, unit='s').strftime('%Y-%m-%d')
    
    filtered_inventory = historial_inventario_df[
        (historial_inventario_df['Tienda'] == selected_store) & 
        (historial_inventario_df['Fecha'] == selected_date)
    ]
    
    filtered_demand = demandaEsperada[
        (demandaEsperada['Tienda'] == selected_store) & 
        (demandaEsperada['Fecha'] == selected_date)
    ]
    
    # Merge the dataframes on SKU to calculate the difference
    merged_df = pd.merge(filtered_inventory, filtered_demand, on=['SKU', 'Tienda'], how='inner')
    merged_df['Diferencia'] = merged_df['Inventario Antes'] - merged_df['Demanda Esperada']
    total_difference = merged_df['Diferencia'].sum()
    
    fig = go.Figure(data=[
        go.Bar(name='Inventario', x=filtered_inventory['SKU'], y=filtered_inventory['Inventario Antes']),
        go.Bar(name='Demanda Esperada', x=filtered_demand['SKU'], y=filtered_demand['Demanda Esperada'])
    ])
    
    fig.update_layout(barmode='group', title=f'Inventario y Demanda Esperada para {selected_store} el {selected_date}')
    
    difference_marker = f'Suma de Diferencia (Inventario - Demanda Esperada): {total_difference}'
    
    return fig, difference_marker

# Run the app
if __name__ == '__main__':
    app.run_server(debug=False)
```

###### Animacion

Hacemos una grafica

```python
# Suponiendo que demandaEsperada y historial_inventario_df ya están definidos

# Filtrar los dataframes
fecha_filtro = '2024-05-30'
tienda_filtro = 'Tienda_6_Marca_8'

demanda_filtrada = demandaEsperada[(demandaEsperada['Fecha'] == fecha_filtro) & (demandaEsperada['Tienda'] == tienda_filtro)]
inventario_filtrado = historial_inventario_df[(historial_inventario_df['Fecha'] == fecha_filtro) & (historial_inventario_df['Tienda'] == tienda_filtro)]

# Unir los dataframes filtrados en base al SKU
df_combinado = pd.merge(demanda_filtrada, inventario_filtrado, on=['Fecha', 'SKU', 'Tienda'])

# Calcular la diferencia
df_combinado['Diferencia'] = df_combinado['Inventario Antes'] - df_combinado['Demanda Esperada']
suma_diferencia = df_combinado['Diferencia'].sum()

# Crear la gráfica
plt.figure(figsize=(20, 12))
plt.bar(df_combinado['SKU'], df_combinado['Inventario Antes'], width=0.4, label='Inventario', align='center', color='#006b60')
plt.bar(df_combinado['SKU'], df_combinado['Demanda Esperada'], width=0.4, label='Demanda Esperada', align='edge', color='#ffc42a')

plt.xlabel('SKU')
plt.ylabel('Cantidad')
plt.title(f'Inventario y Demanda Esperada para {tienda_filtro} en {fecha_filtro}')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()

# Añadir la suma de la diferencia en el título o como anotación en la gráfica
plt.figtext(0.5, 0.95, f'Suma de Diferencia (Inventario - Demanda Esperada): {suma_diferencia}', 
            horizontalalignment='center', fontsize=20, color='#3c6997', weight='bold')

plt.show()
```

Hacemos la generacion de todos los dias para una tienda

```python
# Suponiendo que demandaEsperada y historial_inventario_df ya están definidos

# Filtro de la tienda
tienda_filtro = 'Tienda_6_Marca_8'

# Crear un directorio para almacenar las imágenes
output_dir = '/home/ricmunrom/Documents/jorgeLozano/EUgeNia/primeraExploracion/outputsModelo/imagenesAnimacionOptimizada/'
os.makedirs(output_dir, exist_ok=True)

# Fechas de inicio y fin
start_date = datetime.strptime('2024-05-20', '%Y-%m-%d')
end_date = datetime.strptime('2024-06-20', '%Y-%m-%d')
delta = timedelta(days=1)

current_date = start_date
while current_date <= end_date:
    fecha_filtro = current_date.strftime('%Y-%m-%d')
    
    demanda_filtrada = demandaEsperada[(demandaEsperada['Fecha'] == fecha_filtro) & (demandaEsperada['Tienda'] == tienda_filtro)]
    inventario_filtrado = historial_inventario_df[(historial_inventario_df['Fecha'] == fecha_filtro) & (historial_inventario_df['Tienda'] == tienda_filtro)]
    
    df_combinado = pd.merge(demanda_filtrada, inventario_filtrado, on=['Fecha', 'SKU', 'Tienda'])
    
    # Calcular la diferencia
    df_combinado['Diferencia'] = df_combinado['Inventario Antes'] - df_combinado['Demanda Esperada']
    suma_diferencia = df_combinado['Diferencia'].sum()
    
    # Crear la gráfica
    plt.figure(figsize=(20, 12))
    plt.bar(df_combinado['SKU'], df_combinado['Inventario Antes'], width=0.4, label='Inventario', align='center', color='#006b60')
    plt.bar(df_combinado['SKU'], df_combinado['Demanda Esperada'], width=0.4, label='Demanda Esperada', align='edge', color='#ffc42a')
    
    plt.xlabel('SKU')
    plt.ylabel('Cantidad')
    plt.title(f'Inventario y Demanda Esperada para {tienda_filtro} en {fecha_filtro}')
    plt.xticks(rotation=90)
    plt.legend()
    plt.tight_layout()
    
    # Añadir la suma de la diferencia en el título o como anotación en la gráfica
    plt.figtext(0.5, 0.95, f'Suma de Diferencia (Inventario - Demanda Esperada): {suma_diferencia}', 
                horizontalalignment='center', fontsize=20, color='#3c6997', weight='bold')
    
    # Guardar la gráfica
    output_path = os.path.join(output_dir, f'graph_{fecha_filtro}.png')
    plt.savefig(output_path)
    plt.close()
    
    current_date += delta
```
